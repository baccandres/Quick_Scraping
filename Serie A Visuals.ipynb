```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_standings = pd.read_csv('Standings.csv')
df_standings = df_standings.drop(['Unnamed: 0'], axis=1)
df_standings = df_standings[df_standings.Campaign < 2021]
df_standings
```

```python
len(df_standings.Teams.unique())
```

```python
df_winners = df_standings.loc[df_standings.groupby('Campaign')['P'].idxmax()].set_index('Campaign')
sns.countplot(x='Teams', data=df_winners,
             facecolor=(0,0,0,0), linewidth=5,
             edgecolor=sns.color_palette("dark", 3));
plt.savefig('images/championships.jpg')
```

```python
pd.DataFrame(df_winners.Teams.value_counts())
```

```python
df_top3 = df_standings.loc[(df_standings.Teams == 'JUV') | (df_standings.Teams == 'INT') | (df_standings.Teams == 'MIL')]
sns.lineplot(data=df_top3, x='Campaign', y='P', hue='Teams');
plt.savefig('images/top3_performance.jpg')
```

```python
round(df_top3.groupby('Teams').agg({'P':'mean'}),2)
```

```python
sns.regplot(x='F', y='P', data=df_standings);
plt.savefig('images/reg_goals.jpg')
```

```python
round(np.corrcoef(df_standings.F, df_standings.P)[0][1]**2,2)
```

```python
sns.regplot(x='A', y='P', data=df_standings);
plt.savefig('images/reg_against.jpg')
```

```python
round(np.corrcoef(df_standings.A, df_standings.P)[0][1]**2,2)
```

```python
df_mostGoals = df_standings.loc[df_standings.groupby('Campaign')['F'].idxmax()].set_index('Campaign')
sns.countplot(x='Teams', data=df_mostGoals,
             facecolor=(0,0,0,0), linewidth=5,
             edgecolor=sns.color_palette("dark", 3));
plt.savefig('images/mostgoals.jpg')
```

```python
df_leastConceded = df_standings.loc[df_standings.groupby('Campaign')['A'].idxmin()].set_index('Campaign')
sns.countplot(x='Teams', data=df_leastConceded,
             facecolor=(0,0,0,0), linewidth=5,
             edgecolor=sns.color_palette("dark", 3));
plt.savefig('images/leastgoals.jpg')
```

```python
sns.scatterplot(x='A', y='F', data=df_winners, hue='Teams');
plt.savefig('images/reg_top3.jpg')
```
